```python
from datetime import datetime

now = datetime.now()
date = now.strftime("%Y/%m/%d")

import openpyxl

workbook = openpyxl.Workbook()
ws = workbook.active

ws['B2'].value = '請求書'
ws['B4'].value = '株式会社ABC'
ws['B5'].value = '〒101-0022 東京都千代田区神田練塀町300'
ws['B6'].value = 'TEL:03-1234-5678 FAX:03-1234-5678'
ws['B7'].value = '担当者名:鈴木一郎 様'

ws['F4'].value = 'No.'
ws['G4'].value = "0001"
ws['F5'].value = '日付'
ws['G5'].value = date

header = ['商品名', '数量', '単価', '金額']

target_row = ws.max_row + 3

for i, value in enumerate(header, start=2):
    ws.cell(row=target_row, column=i).value = value

data = [
    ['商品A', 2, 10000, 20000],
    ['商品B', 1, 15000, 15000]
]

target_row_2 = ws.max_row + 1

for row_idx, row_data in enumerate(data):
    current_row = target_row_2 + row_idx
    for col_idx, val in enumerate(row_data, start=2):
        ws.cell(row=current_row, column=col_idx).value = val

subtotal = 35000
tax = int(subtotal * 0.1)
total = subtotal + tax

ws["E13"] .value= subtotal

title = ['小計', '消費税', '税込合計']

target_row_3 = ws.max_row + 2

for i, value in enumerate(title):
    ws.cell(row=target_row_3 + i, column=2).value = value

ws['E15'].value = subtotal
ws['E16'].value = tax
ws['E17'].value = total

workbook.save('kadai_001.xlsx')

```
