```python
import pandas as pd
from openpyxl.styles import PatternFill

file_list = ['2022_年間売上表.xlsx', '2023_年間売上表.xlsx']

df = pd.DataFrame()

for file_name in file_list:
    temp_df = pd.read_excel(file_name)
    temp_df.columns = ['売上年', '商品', '金額（千円）']
    df = pd.concat([df, temp_df], ignore_index=True)

result = df.groupby(['商品', '売上年'], as_index=False)['金額（千円）'].sum()

writer = pd.ExcelWriter('売上集計表.xlsx')

result.to_excel(writer, sheet_name='Sheet1', index=False)

workbook = writer.book
sheet = writer.sheets['Sheet1']

header_fill = PatternFill(patternType='solid', fgColor='F2F2F2')

for cell in ['A1', 'B1', 'C1']:
    sheet[cell].fill = header_fill

writer.close()
```
